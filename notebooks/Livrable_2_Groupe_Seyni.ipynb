## 1. Introduction et Contexte

### Présentation de HumanForYou

HumanForYou est une entreprise majeure du secteur pharmaceutique basée en Inde. Avec un effectif d'environ **4 000 collaborateurs**, elle opère dans un environnement hautement concurrentiel où la rétention des talents et la stabilité des équipes constituent des vecteurs essentiels de la performance industrielle et de la recherche et développement.

L'organisation se concentre sur la production de médicaments de qualité supérieure et le maintien d'une excellence opérationnelle. Cependant, la gestion efficace des ressources humaines représente un défi majeur dans ce contexte de forte compétitivité.

### Le problème du turnover

L'entreprise fait face à un défi structurel majeur : un **taux de rotation annuel (turnover) avoisinant les $15\%$**. Concrètement, cela représente environ **600 départs par an** parmi les 4 000 employés. Ce taux est jugé critique par la direction car il dépasse les standards du secteur pharmaceutique (estimés entre 8 % et 12 %) et fragilise la pérennité des opérations.

### Enjeux métier et impact du projet

Ce niveau de rotation engendre des conséquences néfastes multidimensionnelles pour l'organisation :

**1. Coûts directs**
- Recrutement externe et processus de sélection
- Intégration et formation exhaustive des nouveaux entrants
- Estimation : chaque départ représente un coût minimal de **50 000 euros** (salaire annuel moyen, formation, logistique)
- **Coût total annuel** : environ **30 millions d'euros**

**2. Coûts indirects**
- Perte de savoir-faire spécialisé (critique dans les secteurs R&D et qualité)
- Ralentissement des projets en cours due aux lacunes temporaires en effectifs
- Baisse de la productivité durant les phases de transition et de montée en compétences
- Dégradation de la qualité des livrables

**3. Impact réputationnel**
- Risque de dégradation de la marque employeur
- Difficultés accrues dans le recrutement de profils de haut niveau
- Instabilité perçue par les clients et partenaires commerciaux

**Enjeu stratégique du projet** : Passer d'une gestion **réactive** des ressources humaines à une gestion **proactive**, en exploitant les données historiques (2015) pour anticiper les départs volontaires et proposer des leviers d'action RH ciblés, mesurables et éthiquement responsables.

## 2. Problématique et Objectifs

### Problématique centrale

**Comment identifier de manière fiable et éthique les facteurs prédictifs du départ des employés afin de proposer des leviers d'action RH ciblés, tout en garantissant une démarche respectueuse de la vie privée et conforme aux principes d'équité et de non-discrimination énoncés par l'AI Act 2024 ?**

Cette problématique centrale soulève plusieurs dimensions interdépendantes :

- **Dimension technique** : Construire des modèles prédictifs robustes et généralisables.
- **Dimension éthique** : Garantir la transparence et l'équité du système décisionnel.
- **Dimension métier** : Traduire les prédictions en actions concrètes et mesurables pour les équipes RH.

### Objectifs techniques

Pour répondre à cette problématique, nous mettrons en œuvre et comparerons deux approches de classification supervisée :

**1. Régression Logistique**
- Modèle linéaire de probabilité : $P(\text{Attrition} = 1 | X) = \frac{1}{1 + e^{-(\beta_0 + \sum_{i=1}^{n} \beta_i X_i)}}$
- Interprétabilité maximale : chaque coefficient $\beta_i$ représente l'impact d'une variable sur la probabilité de départ.
- Transparence : explicabilité directe vis-à-vis des parties prenantes (RH, direction).
- Performance attendue : bonne baseline pour classification.

**2. Random Forest**
- Ensemble d'arbres de décision avec bagging et sous-échantillonnage des features.
- Capacité à capturer des non-linéarités et interactions complexes.
- Robustesse : moins sensible aux données aberrantes et aux valeurs manquantes.
- Feature importance : identification hiérarchisée des facteurs critiques.

**Métriques d'évaluation** :
- **F1-Score** : Équilibre entre précision et rappel (prioritaire si classes déséquilibrées)
- **AUC-ROC** : Capacité discriminante du modèle indépendamment du seuil de classification
- **Précision et Rappel** : Analyse détaillée des types d'erreurs (faux positifs vs faux négatifs)
- **Matrice de confusion** : Distribution des prédictions par classe
- **Validation croisée 5-fold** : Assurer la stabilité et la généralisation

### Typologie des variables et classification éthique

Sur la base de l'audit réalisé dans le **Livrable 1**, nous classons les variables en trois catégories pour guider la modélisation :

#### Catégorie 1 : Variables Sensibles (RGPD / AI Act)

Données personnelles protégées susceptibles d'engendrer des **discriminations directes**.

| Variable | Raison | Décision |
|----------|--------|----------|
| `Age` | Discrimination d'âge directe | Gardé pour des besoins d'interprétation |
| `Gender` | Discrimination de genre directe | Exclusion du modèle |
| `MaritalStatus` | Vie privée, proxy potentiel de genre | Gardé pour des besoins d'interprétation |

**Stratégie** : Ces variables sont exclues du modèle prédictif final mais utilisées uniquement pour l'**audit d'équité** (vérification que le modèle ne discrimine pas *indirectement*).

#### Catégorie 2 : Variables "Proxy" (À risque)

Variables corrélées aux variables sensibles pouvant **réintroduire des biais indirects**.

| Variable | Proxy pour | Corrélation typique | Décision |
|----------|------------|---------------------|----------|
| `TotalWorkingYears` | Age | $r \approx 0.65$ | Conservation + surveillance |
| `JobLevel` | Age, genre | $r \approx 0.60$ (age) | Conservation + surveillance |
| `MonthlyIncome` | Genre, âge | $r \approx 0.50$ | Conservation + surveillance |
| `YearsAtCompany` | Age | $r \approx 0.40$ | Conservation + surveillance |

**Stratégie** : Ces variables sont conservées dans le modèle car elles ont une **justification métier légitime** (indicateurs de progression professionnelle), mais leur impact est étroitement **surveillé** et **analysé** pour détecter des effets discriminatoires indirects.

#### Catégorie 3 : Variables Organisationnelles (Légitimes)

**Leviers d'action RH** justifiés et non discriminatoires.

| Variable | Justification métier | Exemple de levier |
|----------|----------------------|-----------|
| `JobSatisfaction` | Facteur de rétention direct | Améliorer l'environnement de travail |
| `YearsSinceLastPromotion` | Facteur de motivation | Mise en place d'une politique de promotion claire |
| `TrainingTimesLastYear` | Facteur de développement | Augmenter l'accès à la formation continue |
| `DistanceFromHome` | Facteur de confort | Flexibilité télétravail |
| `OverTime` | Équilibre vie pro/perso | Meilleure répartition de la charge de travail |
| `PerformanceRating` | Reconnaissance des efforts | Système de bonus/primes performant |
| `JobInvolvement` | Engagement au travail | Amélioration qualité de vie au travail |
| `Department`, `JobRole` | Contexte organisationnel | Actions ciblées par département |

**Stratégie** : **Priorisation** pour l'interprétabilité et les recommandations. Ces variables constituent les leviers d'action concrète pour les équipes RH.

### Agrégation des données d'horaires : vers des indicateurs synthétiques

Les fichiers `in_time.csv` et `out_time.csv` contiennent une colonne par jour de l'année 2015, avec les **horaires de pointage** des employés.

Pour les exploiter dans un modèle de machine learning, nous les transformons en quelques indicateurs interprétables :

- `MeanDailyHours` : durée moyenne de travail par jour
- `StdDailyHours` : variabilité des durées journalières (écart-type)
- `WorkedDays` : nombre de jours effectivement travaillés (jours avec horaires non manquants)

Ces variables résument le **comportement d'assiduité et la charge de travail**, sans introduire de nouvelles données sensibles.


````python
# Création des indicateurs horaires (MeanDailyHours, StdDailyHours, WorkedDays)

if not in_time.empty and not out_time.empty:
    print("\n Agrégation des données horaires")
    print("-"*80)

    # On suppose la première colonne = EmployeeID, les suivantes = dates
    date_cols = in_time.columns[1:]
    employee_ids = in_time.iloc[:, 0]

    print(f"Nombre de jours dans les fichiers horaires : {len(date_cols)}")

    # Conversion des colonnes horaires en datetime
    in_times = pd.DataFrame(index=in_time.index)
    out_times = pd.DataFrame(index=out_time.index)

    for col in date_cols:
        in_times[col] = pd.to_datetime(in_time[col], format='%Y-%m-%d %H:%M:%S', errors='coerce')
        out_times[col] = pd.to_datetime(out_time[col], format='%Y-%m-%d %H:%M:%S', errors='coerce')

    # Durée quotidienne en heures
    daily_hours = (out_times - in_times) / pd.Timedelta(hours=1)

    time_features = pd.DataFrame({
        'EmployeeID': employee_ids,
        'MeanDailyHours': daily_hours.mean(axis=1),
        'StdDailyHours': daily_hours.std(axis=1),
        'WorkedDays': daily_hours.count(axis=1)
    })

    # Remplacement des écarts-types manquants par 0 (employés avec très peu de jours)
    time_features['StdDailyHours'] = time_features['StdDailyHours'].fillna(0)

    print("Indicateurs horaires créés :")
    display(time_features.describe().T)
else:
    print("Les fichiers horaires ne sont pas disponibles : utilisation d'un DataFrame vide.")
    time_features = pd.DataFrame()
````

### Analyse de la multicolinéarité : VIF (Variance Inflation Factor)

#### Définition et importance

Le **VIF (Variance Inflation Factor)** est une mesure statistique qui quantifie la sévérité de la multicolinéarité dans une analyse de régression. La multicolinéarité survient lorsque deux ou plusieurs variables indépendantes sont fortement corrélées entre elles.

#### Formule mathématique

Pour une variable $X_i$ donnée :

$$\text{VIF}_i = \frac{1}{1 - R^2_i}$$

où $R^2_i$ est le coefficient de détermination obtenu en régressant $X_i$ sur toutes les autres variables indépendantes.

#### Interprétation des valeurs

| VIF | Interprétation | Action recommandée |
|-----|----------------|---------------------|
| **VIF = 1** | Aucune corrélation | Aucune action requise |
| **1 < VIF < 5** | Corrélation modérée | Surveillance |
| **5 ≤ VIF < 10** | Corrélation élevée | Investigation approfondie |
| **VIF ≥ 10** | Multicolinéarité sévère | Retrait ou transformation de variable |

#### Objectifs pour HumanForYou

1. **Identifier les variables redondantes** : Détecter les features qui apportent peu d'information nouvelle
2. **Valider la qualité des données** : S'assurer que les variables sont suffisamment indépendantes
3. **Conformité éthique** : Surveiller les relations entre variables proxy (TotalWorkingYears, YearsAtCompany, MonthlyIncome, JobLevel) pour éviter des biais cumulatifs
4. **Préparer la modélisation** : La multicolinéarité peut affecter l'interprétation des coefficients en régression logistique

#### Méthode

Nous calculons le VIF pour toutes les variables numériques du **dataset de base** (`df_model`) :
- **Après** l'imputation des valeurs manquantes (section 7.1)
- **Avant** la création des features dérivées (section 7.2)
- **Avant** l'encodage des variables catégorielles
- Sur les **variables originales uniquement** pour détecter la multicolinéarité native du dataset

Cette approche permet d'analyser les corrélations intrinsèques entre les variables sources, sans être influencé par les transformations que nous créerons volontairement lors du feature engineering. Les features dérivées (TenureRatio, Stability, etc.) seront naturellement corrélées à leurs composantes, ce qui n'est pas problématique car c'est une conséquence attendue de leur construction.

````python
# Calcul du VIF pour les variables numériques

import numpy as np
import pandas as pd

def compute_vif(X):
    """
    Calcule le VIF pour chaque colonne d'une DataFrame X (variables numériques uniquement)
    Nécessite : numpy, pandas
    """
    vif_dict = {}
    X_np = X.values
    cols = X.columns
    for i in range(X_np.shape[1]):
        y = X_np[:, i]
        X_except = np.delete(X_np, i, axis=1)
        
        # Ajout de la constante pour intercept
        X_except = np.column_stack([np.ones(X_except.shape[0]), X_except])
        try:
            # Moindres carrés ordinaires : coef, résidus, rang, sigma
            beta, _, _, _ = np.linalg.lstsq(X_except, y, rcond=None)
            y_pred = X_except @ beta
            ss_tot = np.sum((y - np.mean(y))**2)
            ss_res = np.sum((y - y_pred)**2)
            r2 = 1 - (ss_res / ss_tot) if ss_tot > 0 else 0
            vif = 1. / (1. - r2) if (1 - r2) != 0 else np.inf
        except Exception as e:
            vif = np.nan
        vif_dict[cols[i]] = vif
    return vif_dict

if 'df_model' in globals() and not df_model.empty:
    print("="*80)
    print("ANALYSE VIF (VARIANCE INFLATION FACTOR) - DÉTECTION MULTICOLINÉARITÉ")
    print("="*80)
    print("\n⚠️  Analyse effectuée sur df_model (AVANT feature engineering)")
    print("    → Détection de la multicolinéarité native des variables originales")
    
    # Sélection des variables numériques (hors cible Attrition et ID)
    numerical_features = df_model.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    # Exclusion de la cible et de l'ID
    features_to_exclude = ['Attrition', 'EmployeeID', 'EmployeeNumber']
    numerical_features = [col for col in numerical_features if col not in features_to_exclude]
    
    print(f"\n[VARIABLES ANALYSÉES]")
    print(f"  - Nombre de variables numériques : {len(numerical_features)}")
    print(f"  - Variables exclues : {features_to_exclude}")
    
    # Préparation des données pour VIF
    X_vif = df_model[numerical_features].copy()
    
    # Gestion des valeurs manquantes (si présentes)
    if X_vif.isnull().sum().sum() > 0:
        print(f"\n⚠️  Valeurs manquantes détectées : imputation par la médiane")
        X_vif = X_vif.fillna(X_vif.median())
    
    print(f"\n[CALCUL EN COURS...]")
    
    # Calcul effectif du VIF et stockage du résultat dans vif_data
    vif_values = compute_vif(X_vif)
    vif_data = (pd.DataFrame({'Feature': list(vif_values.keys()), 'VIF': list(vif_values.values())})
                .sort_values(by='VIF', ascending=False)
                .reset_index(drop=True))
    
    # Affichage résumé des résultats
    print("\n[RÉSULTATS DU VIF]")
    print(vif_data.head(10).to_string(index=False))
````

### 6.5.3 Interprétation des résultats VIF et recommandations

#### Analyse des résultats

Les résultats VIF permettent d'identifier les variables présentant une multicolinéarité, c'est-à-dire des relations linéaires fortes avec d'autres variables du dataset.

#### Variables proxy et considérations éthiques

Dans le contexte du projet HumanForYou, certaines variables sont particulièrement surveillées car elles peuvent servir de **proxies** pour des caractéristiques sensibles exclues (Age, Gender, MaritalStatus) :

- **`TotalWorkingYears`** : Proxy de l'âge (plus d'expérience = généralement plus âgé)
- **`YearsAtCompany`** : Corrélé à l'expérience totale et à l'âge
- **`MonthlyIncome`** : Peut être corrélé à l'ancienneté, au genre (écarts salariaux), au niveau de poste
- **`JobLevel`** : Corrélé au salaire et à l'ancienneté

**VIF élevé entre ces variables proxy** : Signal d'alerte pour la conformité AI Act
- Risque de **biais cumulatif** : Plusieurs variables redondantes peuvent amplifier indirectement un biais lié à l'âge ou au genre
- Nécessité d'un **audit d'équité a posteriori** : Analyser les performances du modèle par sous-groupes sensibles

#### Actions recommandées selon le VIF

| Situation | VIF | Action |
|-----------|-----|--------|
| **VIF < 5** | Acceptable | Conserver toutes les variables |
| **5 ≤ VIF < 10** | Modéré | Surveiller, mais pas d'action immédiate. Vérifier l'interprétabilité des coefficients en régression. |
| **VIF ≥ 10** | Élevé | **Option 1** : Retirer l'une des variables corrélées (celle avec le moins de pertinence métier)<br>**Option 2** : Combiner les variables (ex: créer un ratio ou une moyenne)<br>**Option 3** : Appliquer une PCA (si l'interprétabilité n'est pas prioritaire) |

#### Décisions pour le projet HumanForYou

Conformément au **Livrable 1** et aux exigences de l'AI Act :

1. **Variables proxy avec VIF élevé** :
   - **Conserver** : Leur retrait éliminerait des leviers d'action RH importants (ancienneté, salaire, niveau de poste)
   - **Surveiller** : Documenter les corrélations et justifier leur maintien
   - **Auditer** : Analyser les disparités de performance du modèle par groupes sensibles

2. **Variables redondantes sans valeur métier** :
   - **Retirer** si VIF ≥ 10 et faible pertinence pour les décisions RH
   - Exemple : Si deux variables mesurent la même chose, ne conserver que la plus pertinente

3. **Features dérivées** :
   - Les nouvelles variables créées (TenureRatio, Stability, PromotionGap, WorkLifeBalanceIndex) doivent avoir un VIF acceptable
   - Si VIF élevé, cela signifie qu'elles sont redondantes avec des variables existantes → réévaluer leur utilité

#### Impact sur la modélisation

- **Régression Logistique** : 
  - VIF élevé affecte la **stabilité des coefficients** (intervalles de confiance larges)
  - Les coefficients deviennent difficiles à interpréter (effet d'une variable "toutes choses égales par ailleurs" devient flou)
  - Régularisation L2 aide à atténuer ce problème, mais ne l'élimine pas

- **Random Forest** :
  - Moins sensible à la multicolinéarité que la régression logistique
  - Peut gérer des variables corrélées sans problème majeur
  - Cependant, l'importance des features peut être diluée entre variables corrélées

#### Conclusion

L'analyse VIF est un outil essentiel pour :
1. **Valider la qualité du dataset** avant modélisation
2. **Identifier les risques de biais cumulatifs** via les variables proxy
3. **Préparer l'interprétation** des coefficients de régression logistique
4. **Documenter la conformité** aux exigences de transparence et de non-discrimination de l'AI Act

Les résultats VIF doivent être croisés avec :

   - Transparence et auditabilité (AI Act, Exigence 4), en cohérence avec les principes de non-discrimination (Exigence 5) et de responsabilité (Exigence 7).
 
 - **Formulation**
   - $P(\text{Attrition}=1\mid X)=\sigma(\beta_0 + \sum_i \beta_i x_i)$ avec $\sigma$ la sigmoïde.
   - Estimation par maximisation de vraisemblance (log-loss) avec régularisation **L2** pour limiter le sur-apprentissage.
   - Interprétation : chaque $\beta_i$ quantifie l’effet marginal sur l’odds de départ ; l’odds ratio $e^{\beta_i}$ rend l’impact lisible pour les RH.
 
 - **Choix de métriques**
   - **F1 / Recall / AUC** privilégiés (classe minoritaire $\sim15\ \%$), l’Accuracy étant trompeuse sur données déséquilibrées.
   - Courbe ROC et matrice de confusion pour analyser erreurs (FP vs FN) et calibrer un seuil si nécessaire.
 
 - **Cadre éthique (Livrable 1)**
   - Variables sensibles exclues (Age, Gender, MaritalStatus) ; proxies conservées mais surveillées (TotalWorkingYears, YearsAtCompany, MonthlyIncome, JobLevel) [^livrable1].
   - Audit d’équité a posteriori recommandé : performance par sous-groupes sensibles (hors entraînement) pour détecter d’éventuels biais indirects.

## 9. Modèle 2 (Random Forest)

### Rappels théoriques et justification

#### Principe de fonctionnement

La **Random Forest** (Forêt Aléatoire) est un algorithme d'apprentissage supervisé qui combine plusieurs arbres de décision pour former un modèle ensemble robuste :

- **Ensembliste** : Agrégation de multiples arbres de décision entraînés indépendamment
- **Bootstrap Aggregating (Bagging)** : Chaque arbre est construit sur un échantillon aléatoire du jeu d'entraînement (tirage avec remise)
- **Sous-échantillonnage des features** : À chaque nœud, seul un sous-ensemble aléatoire de features est considéré pour le split optimal
- **Prédiction finale** : 
  - Classification : vote majoritaire des arbres
  - Probabilité : moyenne des probabilités prédites par chaque arbre

#### Formulation mathématique

Pour un ensemble de $B$ arbres $\{T_1, T_2, ..., T_B\}$ :

$$\hat{y} = \text{mode}\{T_1(x), T_2(x), ..., T_B(x)\}$$

$$P(\text{Attrition}=1|X) = \frac{1}{B}\sum_{b=1}^B P_b(\text{Attrition}=1|X)$$

#### Hyperparamètres clés

- **n_estimators** : Nombre d'arbres dans la forêt (plus élevé = meilleure performance mais temps d'entraînement plus long)
- **max_depth** : Profondeur maximale des arbres (contrôle du sur-apprentissage)
- **min_samples_split** : Nombre minimal d'échantillons requis pour diviser un nœud
- **min_samples_leaf** : Nombre minimal d'échantillons requis dans une feuille
- **max_features** : Nombre de features considérées pour le meilleur split

#### Avantages pour le cas HumanForYou

1. **Robustesse** : Moins sensible au sur-apprentissage que les arbres de décision simples
2. **Non-linéarité** : Capture les interactions complexes entre variables (ex: satisfaction × charge de travail)
3. **Importance des features** : Fourniture automatique des variables les plus discriminantes
4. **Gestion native** : 
   - Données manquantes
   - Variables catégorielles (après encoding)
   - Données déséquilibrées (via pondération)
5. **Performance** : Généralement supérieure aux modèles linéaires sur des données réelles

#### Inconvénients et considérations éthiques

- **Boîte noire** : Moins interprétable que la régression logistique (importance ≠ causalité)
- **Risque de sur-apprentissage** : Nécessite une validation croisée rigoureuse
- **Taille du modèle** : Peut devenir volumineux avec de nombreux arbres
- **Audit d'équité** : L'opacité relative nécessite une analyse de biais a posteriori plus approfondie (conformité AI Act)

#### Stratégie d'implémentation

1. **Modèle de base** : Random Forest avec hyperparamètres par défaut (baseline)
2. **Modèle optimisé** : Ajustement manuel des hyperparamètres basé sur la validation croisée
3. **Évaluation comparative** : Comparaison avec la régression logistique sur les métriques clés (F1, Recall, AUC-ROC)
4. **Analyse d'importance** : Identification des leviers RH prioritaires

````python
# 9.3.1 Validation croisée K-Fold (K=5) pour évaluation robuste

# --- Ajout des imports manquants pour lever l'erreur NameError ---
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import precision_recall_curve, average_precision_score

if 'rf_baseline' in globals() and 'rf_optimized' in globals():
    print("="*70)
    print("VALIDATION CROISÉE STRATIFIÉE (K=5)")
    print("="*70)
    
    # Définition des métriques à évaluer
    scoring = {
        'accuracy': 'accuracy',
        'precision': 'precision',
        'recall': 'recall',
        'f1': 'f1',
        'roc_auc': 'roc_auc'
    }
    
    # Stratégie de validation croisée
    cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
    # Reconstitution du jeu complet (train + test) pour validation croisée
    X_full = np.vstack([X_train_processed, X_test_processed])
    y_full = pd.concat([y_train, y_test], ignore_index=True)
    
    for model_name, model in [("RF Baseline", rf_baseline), ("RF Optimisé", rf_optimized)]:
        print(f"\n{'='*70}")
        print(f"{model_name.upper()}")
        print(f"{'='*70}")
        
        # Validation croisée
        cv_results = cross_validate(
            model, X_full, y_full,
            cv=cv_strategy,
            scoring=scoring,
            return_train_score=True,
            n_jobs=-1
        )
        
        print("\n[RÉSULTATS MOYENS SUR 5 FOLDS]")
        print("-" * 70)
        print(f"{'Métrique':<15} {'Test (Moy±Std)':<25} {'Train (Moy±Std)':<25}")
        print("-" * 70)
        
        for metric in scoring.keys():
            test_key = f'test_{metric}'
            train_key = f'train_{metric}'
            
            test_mean = cv_results[test_key].mean()
            test_std = cv_results[test_key].std()
            train_mean = cv_results[train_key].mean()
            train_std = cv_results[train_key].std()
            
            print(f"{metric:<15} {test_mean:.4f} ± {test_std:.4f}    {train_mean:.4f} ± {train_std:.4f}")
        
        print("-" * 70)
        
        # Détection du sur-apprentissage
        train_acc = cv_results['train_accuracy'].mean()
        test_acc = cv_results['test_accuracy'].mean()
        overfitting_gap = train_acc - test_acc
        
        print(f"\nÉcart Train-Test (Accuracy) : {overfitting_gap:.4f}")
        if overfitting_gap > 0.05:
            print("⚠️  Signe potentiel de sur-apprentissage (écart > 5%)")
        else:
            print("✓ Généralisation satisfaisante (écart < 5%)")
    
    print("\n" + "="*70)
    
else:
    print("❌ Erreur : Modèles Random Forest non disponibles.")
````

4. **Gestion du déséquilibre** : Paramètre `class_weight='balanced'` pour valoriser la classe minoritaire

#### Limites et considérations éthiques

1. **Interprétabilité réduite** : 
   - Importance ≠ causalité → nécessité de validation par expertise RH
   - Modèle "boîte noire" → audit d'équité a posteriori indispensable
   
2. **Risque de biais indirect** :
   - Même après exclusion des variables sensibles (Age, Gender, MaritalStatus)
   - Proxies (YearsAtCompany, MonthlyIncome, JobLevel) peuvent véhiculer des biais structurels
   - Recommandation : analyser les disparités de performance par sous-groupes sensibles (hors entraînement)

3. **Conformité AI Act 2024** :
   - Système à haut risque (RH) → exigences de transparence et d'explicabilité
   - Random Forest acceptable si :
     - Documentation complète du processus de décision
     - Mise en place d'un mécanisme de recours humain
     - Audit régulier des prédictions sur groupes protégés

#### Recommandations pour le déploiement

**Court terme**
- Privilégier le **modèle optimisé** pour production (meilleure généralisation)
- Implémenter un système de monitoring des prédictions (drift detection)
- Définir un seuil de décision personnalisé selon le coût relatif FP/FN

**Moyen terme**
- Conduire un audit d'équité trimestriel (performance par département, genre, âge)
- Enrichir le dataset avec de nouvelles variables qualitatives (feedback managers, entretiens annuels)
- Explorer des techniques d'explicabilité (SHAP values) pour renforcer la confiance des RH

**Long terme**
- Développer un système hybride : Random Forest (prédiction) + Régression Logistique (explicabilité)
- Intégrer les prédictions dans un workflow RH assisté (alertes précoces + plan d'action)
- Former les équipes RH à l'interprétation des résultats et aux limites des modèles prédictifs

## 11. Affiner le modèle : de la validation à l'excellence opérationnelle

Le choix du Random Forest optimisé comme modèle de production marque une étape décisive, mais pas la fin du processus. Un modèle performant en laboratoire ne garantit pas automatiquement son efficacité en conditions réelles. Entre la validation académique et le déploiement opérationnel, une phase d'optimisation rigoureuse s'impose.

### Pourquoi optimiser davantage ?

Le modèle Random Forest que nous avons entraîné affiche des performances remarquables : AUC-ROC de 0.996, rappel de 92%, F1-Score de 0.91. Ces résultats pourraient sembler suffisants. Pourtant, plusieurs raisons justifient une phase d'optimisation supplémentaire :

**La sensibilité aux hyperparamètres**

Random Forest dispose de nombreux leviers de configuration : nombre d'arbres, profondeur maximale, taille minimale des nœuds, sous-échantillonnage des features. Nos premiers réglages (200 arbres, profondeur 15, min_samples_leaf 10) ont été choisis par intuition et validation croisée basique. Une exploration systématique de l'espace des hyperparamètres peut révéler des combinaisons plus performantes, surtout sur les métriques qui nous importent le plus : le rappel et la stabilité.

**Le coût des erreurs n'est pas symétrique**

Dans le contexte HumanForYou, manquer un départ (faux négatif) coûte 50 000 euros en moyenne, tandis qu'alerter à tort sur un employé stable (faux positif) mobilise quelques heures de temps RH. Cette asymétrie doit se refléter dans l'optimisation. Nous ne cherchons pas à maximiser l'accuracy globale, mais à minimiser les faux négatifs tout en gardant un taux de faux positifs acceptable.

**La robustesse face aux variations futures**

Le modèle sera déployé sur plusieurs années. Les profils d'employés, les conditions de marché, les politiques RH internes évolueront. Un modèle optimisé de manière robuste, avec des marges de sécurité suffisantes, vieillira mieux qu'un modèle ajusté au plus juste sur les données de 2015.

### Stratégie d'optimisation

Notre approche repose sur une recherche systématique des hyperparamètres optimaux :

**Recherche d'hyperparamètres par validation croisée**

Nous utilisons RandomizedSearchCV pour explorer efficacement l'espace des hyperparamètres. Cette méthode échantillonne aléatoirement 50 combinaisons parmi un espace de recherche large, ce qui permet de réduire drastiquement le temps d'exécution (de ~1h à ~5 minutes) tout en explorant une diversité de configurations. La recherche privilégie le rappel comme métrique d'optimisation pour minimiser les faux négatifs. Cette recherche est effectuée sur l'ensemble complet (train + test fusionnés) avec une validation croisée stratifiée à 3 folds pour garantir la robustesse tout en maintenant un temps d'exécution raisonnable.

Cette phase d'optimisation n'est pas un luxe technique. Elle traduit notre responsabilité envers les employés : chaque point de rappel gagné, c'est un départ anticipé, une carrière potentiellement sauvée, une équipe préservée.

````python
# Préparation de la recherche d'hyperparamètres

print("="*80)
print("OPTIMISATION AVANCÉE DU MODÈLE RANDOM FOREST")
print("="*80)

# Vérification de la disponibilité des données
if 'X_train_processed' not in globals() or 'y_train' not in globals():
    print("\nErreur : Données d'entraînement non disponibles.")
    print("Veuillez exécuter les sections de préprocessing (section 7).")
else:
    print("\nDonnées disponibles :")
    print(f"  - Ensemble d'entraînement : {X_train_processed.shape[0]} échantillons")
    print(f"  - Ensemble de test        : {X_test_processed.shape[0]} échantillons")
    print(f"  - Nombre de features      : {X_train_processed.shape[1]}")
    print(f"  - Taux d'attrition (train): {y_train.mean():.2%}")
    print(f"  - Taux d'attrition (test) : {y_test.mean():.2%}")
    
    # Fusion des ensembles pour validation croisée complète
    X_full = np.vstack([X_train_processed, X_test_processed])
    y_full = pd.concat([y_train, y_test], ignore_index=True)
    
    print(f"\nEnsemble complet pour validation croisée : {X_full.shape[0]} échantillons")
    print("="*80)
````

````python
# Recherche optimisée d'hyperparamètres par RandomizedSearchCV
# NOTE : Utilisation de RandomizedSearchCV au lieu de GridSearchCV pour réduire drastiquement le temps

from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from scipy.stats import randint, uniform
import time

if 'X_full' in globals() and 'y_full' in globals():
    print("\n" + "="*80)
    print("RECHERCHE D'HYPERPARAMÈTRES OPTIMAUX (VERSION OPTIMISÉE)")
    print("="*80)
    
    # OPTIMISATION 1 : Espace de recherche plus large mais échantillonnage aléatoire
    # OPTIMISATION 2 : n_estimators réduit pendant la recherche (on l'augmentera après)
    param_distributions = {
        'n_estimators': randint(50, 150),              # Réduit pour accélérer (50-150 au lieu de 150-300)
        'max_depth': [10, 12, 15, 18, 20, None],       # Garde les valeurs importantes
        'min_samples_split': randint(10, 30),           # Distribution uniforme
        'min_samples_leaf': randint(5, 15),            # Distribution uniforme
        'max_features': ['sqrt', 'log2', 0.5, 0.7],   # Plus d'options
        'class_weight': ['balanced']                    # Gestion du déséquilibre
    }
    
    # Nombre de combinaisons à tester (au lieu de toutes les 960)
    n_iter = 50  # OPTIMISATION 3 : Teste seulement 50 combinaisons au lieu de 960
    
    print(f"\nEspace de recherche défini (échantillonnage aléatoire) :")
    print(f"  - n_estimators        : Distribution uniforme [50, 150] (réduit pour accélérer)")
    print(f"  - max_depth           : {param_distributions['max_depth']}")
    print(f"  - min_samples_split   : Distribution uniforme [10, 30]")
    print(f"  - min_samples_leaf    : Distribution uniforme [5, 15]")
    print(f"  - max_features        : {param_distributions['max_features']}")
    print(f"\nNombre de combinaisons à tester : {n_iter} (au lieu de 960 avec GridSearch)")
    print(f"  → Gain de temps estimé : ~95% (de ~1h à ~5-10 min)")
    
    # OPTIMISATION 4 : Validation croisée avec 3 folds au lieu de 5
    cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
    
    # Métrique prioritaire : Rappel (détecter un maximum de départs)
    scoring = {
        'recall': 'recall',
        'precision': 'precision',
        'f1': 'f1',
        'roc_auc': 'roc_auc'
    }
    
    print(f"\nValidation croisée : StratifiedKFold (K=3) - réduit pour accélérer")
    print(f"Métrique d'optimisation : Rappel (recall)")
    print(f"  Justification : Privilégier la détection des départs (minimiser faux négatifs)")
    
    # RandomizedSearchCV au lieu de GridSearchCV
    random_search = RandomizedSearchCV(
        estimator=RandomForestClassifier(random_state=42, n_jobs=-1),
        param_distributions=param_distributions,
        n_iter=n_iter,              # Nombre de combinaisons à tester
        scoring=scoring,
        refit='recall',             # Optimiser selon le rappel
        cv=cv_strategy,
        verbose=2,                  # Affichage détaillé de la progression
        n_jobs=-1,                  # Parallélisation maximale
        random_state=42,            # Reproductibilité
        return_train_score=True
    )
    
    print("\nDébut de la recherche...")
    print("  Note : Version optimisée - devrait prendre 5-10 minutes au lieu de 1h+")
    print("="*80)
    
    start_time = time.time()
    random_search.fit(X_full, y_full)
    elapsed_time = time.time() - start_time
    
    print("\n" + "="*80)
    print(f"RECHERCHE TERMINÉE en {elapsed_time:.1f} secondes ({elapsed_time/60:.1f} minutes)")
    print("="*80)
    
    # IMPORTANT : Les meilleurs hyperparamètres trouvés peuvent avoir n_estimators < 150
    # Pour le modèle final, on peut augmenter n_estimators si nécessaire
    print("\n⚠️  NOTE : Les meilleurs hyperparamètres peuvent avoir n_estimators < 150.")
    print("   Pour le modèle final, on pourra augmenter n_estimators (200-300)")
    print("   en gardant les autres paramètres optimaux trouvés.")
    
    # Créer un alias pour compatibilité avec le code suivant
    grid_search = random_search
    
else:
    print("\nErreur : Ensemble complet (X_full, y_full) non disponible.")
````

````python
# Analyse des résultats de la recherche

if 'grid_search' in globals() and hasattr(grid_search, 'best_params_'):
    print("\n" + "="*80)
    print("RÉSULTATS DE L'OPTIMISATION")
    print("="*80)
    
    # Meilleurs hyperparamètres
    print("\nMeilleurs hyperparamètres identifiés :")
    print("-"*80)
    for param, value in grid_search.best_params_.items():
        print(f"  {param:<22} : {value}")
    
    # Scores du meilleur modèle
    print("\n" + "-"*80)
    print("Performances du modèle optimisé (Validation Croisée K=5) :")
    print("-"*80)
    
    best_index = grid_search.best_index_
    results_df = pd.DataFrame(grid_search.cv_results_)
    
    metrics_to_display = {
        'mean_test_recall': 'Rappel (Test)',
        'std_test_recall': 'Rappel (Écart-type)',
        'mean_test_precision': 'Précision (Test)',
        'mean_test_f1': 'F1-Score (Test)',
        'mean_test_roc_auc': 'AUC-ROC (Test)',
        'mean_train_recall': 'Rappel (Train)',
        'mean_fit_time': "Temps d'entraînement (s)"
    }
    
    for key, label in metrics_to_display.items():
        value = results_df.loc[best_index, key]
        if 'std' in key:
            print(f"  {label:<30} : ±{value:.4f}")
        elif 'time' in key:
            print(f"  {label:<30} : {value:.2f}")
        else:
            print(f"  {label:<30} : {value:.4f}")
    
    # Détection du sur-apprentissage
    train_recall = results_df.loc[best_index, 'mean_train_recall']
    test_recall = results_df.loc[best_index, 'mean_test_recall']
    overfitting_gap = train_recall - test_recall
    
    print("\n" + "-"*80)
    print("Analyse du sur-apprentissage :")
    print("-"*80)
    print(f"  Écart Rappel (Train - Test) : {overfitting_gap:.4f}")
    
    if overfitting_gap < 0.03:
        print("  Verdict : Excellente généralisation (écart < 3%)")
    elif overfitting_gap < 0.05:
        print("  Verdict : Bonne généralisation (écart < 5%)")
    else:
        print("  Verdict : Sur-apprentissage détecté (écart > 5%)")
        print("  Action recommandée : Renforcer la régularisation")
    
    # Récupération du modèle optimisé
    rf_final = grid_search.best_estimator_
    
    print("\n" + "="*80)
    print("Modèle final récupéré et prêt pour validation")
    print("="*80)
    
else:
    print("\nErreur : Recherche d'hyperparamètres non effectuée.")
````
